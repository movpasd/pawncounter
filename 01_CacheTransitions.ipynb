# 01 Cache transitions

Pre-generate chess pawn structure transitions files

```python
import pawncounter
```

```python
# Pre-generate (or load from cache) the transition table for every board size, from 1x2
# up to 4x4. `generate_transitions()` is lazy: it loads data/transitions_WxD.tmp if it
# exists, otherwise runs the BFS and writes it. 4x4 is the heavy one (~2 min, ~2.4 GB).
for width in range(1, 5):
    for depth in range(2, 5):
        board = pawncounter.PawnCounter(width, depth)
        table = board.generate_transitions()
        print(f"{width}x{depth}: {table.height:>12,} transitions")
```
